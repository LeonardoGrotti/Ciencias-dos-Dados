```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python

data = pd.read_excel("PNAD 2013 A.xlsx")
n=data.shape[0]
list(data)
data.head()
```

```python
#Sexo
data.V0302 = data.V0302.astype('category')
data.V0302.cat.categories = ('Masculino','Feminino')

#Raça
data.V0404 = data.V0404.astype('category')
data.V0404.cat.categories = ('Indígena','Branca','Preta','Amarelo','Parda','Sem Declaração')

#Rede
data.V6002 = data.V6002.astype('category')
data.V6002.cat.categories = ('Público','Privado')

#Ensino
data.V4802 = data.V4802.astype('category')
data.V4802.cat.categories = ('Maternal','Alfabetização','Alfabetização Adulta','Primeira à quarta série','Quinta à oitava série','Primeira série regular','Segundo à quinto ano','Sexta à nona série','Educação de jovens','Ensino Médio','Ensino Superior')

#Estados
#data.UF = data.UF.astype('category')
#data.UF.cat.categories = ('Rondonia','Acre','Amazônia','Roraima','Para','Amapa','Tocantins','Maranhao','Piaui','Ceara','RioGrandeN','Paraiba','Pernambuco','Alagoas','Sergipe','Bahia','Minas Gerais','Espirito Santo','Rio de Janeiro','São Paulo','Paraná','Santa Catarina','Rio Grande do Sul','Mato Grosso do Sul','Mato Grosso','Goias','Distrito Federal')

#Curso anterior
data.V0611 = data.V0611.astype('category')
data.V0611.cat.categories = ("Concluiu", 'Não concluiu')

#Ler e Escrever
data.V0601 = data.V0601.astype('category')
data.V0601.cat.categories = ('Sabe ler ou escrever','Não sabe ler ou escrever')


```

```python
regi = pd.DataFrame()
```

```python
regi['REGIAO'] = data.UF
```

```python
regi.REGIAO[(data.UF>=10) & (data.UF<20)] = 1
regi.REGIAO[(data.UF>=20) & (data.UF<30)] = 2
regi.REGIAO[(data.UF>=30) & (data.UF<40)] = 3
regi.REGIAO[(data.UF>=40) & (data.UF<50)] = 4
regi.REGIAO[(data.UF>=50) & (data.UF<60)] = 5

regi.REGIAO.value_counts()
regi.REGIAO = regi.REGIAO.astype('category',ordered = True)
regi.REGIAO.cat.categories = ('NORTE','NORDESTE','SUDESTE','SUL','CENTRO-OESTE')
```

```python
fregiao = (regi.REGIAO.value_counts(True)*100).round(decimals = 1)
plt.ylabel('Valor em %')
fregiao.plot(kind='bar',title = 'Regiões')
```

```python
jt4 = pd.crosstab(regi.REGIAO,data.V6002)
jt4.plot(kind='bar', stacked = True)
```

```python
jt4 = pd.crosstab(regi.REGIAO,data.V4802)
jt4.plot(kind='bar', stacked = True ,figsize = (8,8), color =('salmon','red','blue','green','purple','grey','k','yellow','white','brown','pink'))
```
